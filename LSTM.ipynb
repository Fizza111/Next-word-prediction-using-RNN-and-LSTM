```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts(sentences)
total_words=len(tokenizer.word_index) + 1
print(total_words)
print(tokenizer.word_index)
input_sequences=[]
for line in sentences:
    token_list=tokenizer.texts_to_sequences([line])[0]
    for i in range(1,len(token_list)):
        n_gram_sequences=token_list[:i+1]
        input_sequences.append(n_gram_sequences)
max_sequence_length=max([len(x) for x in input_sequences])
input_sequences=np.array(pad_sequences(input_sequences,maxlen=max_sequence_length,padding="pre"))
print(input_sequences)
```

```python


!pip install wordcloud matplotlib
import matplotlib.pyplot as plt
from wordcloud import WordCloud
# Create a dictionary to hold the word counts for visualization
word_counts = {}
for word, index in tokenizer.word_index.items():
    word_counts[word] = tokenizer.word_counts.get(word, 0) # Use actual counts if available

# Generate a word cloud from the vocabulary
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_counts)
# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Tokenization Vocabulary Word Cloud")
plt.show()
```

```python
#split data into input and labels
X=input_sequences[:, :-1]
print("input data :",X)
y=input_sequences[: ,-1]
y=tf.keras.utils.to_categorical(y,num_classes=total_words)
```

```python
model=Sequential([
    Embedding(total_words,10,input_length=max_sequence_length-1),
    LSTM(80),
    Dense(total_words,activation="softmax")

])
```

```python
model.compile(loss="categorical_crossentropy",optimizer='adam',metrics=['accuracy'])
```

```python


import matplotlib.pyplot as plt
import seaborn as sns


history = model.fit(X, y, epochs=60, verbose=1)
```

```python
# Get training history
acc = history.history['accuracy']
loss = history.history['loss']
epochs = range(1, len(acc) + 1)

# Plot Accuracy
plt.figure(figsize=(5, 3))
sns.lineplot(x=epochs, y=acc, label='Training Accuracy')
plt.title('Training Accuracy over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
# Plot Loss
plt.figure(figsize=(5, 3))
sns.lineplot(x=epochs, y=loss, label='Training Loss', color='salmon')
plt.title('Training Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def predict_next_word(seed_text,next_word=1):
    for _ in range(next_word):
        token_list=tokenizer.texts_to_sequences([seed_text])[0]
        token_list=pad_sequences([token_list],maxlen=max_sequence_length - 1, padding='pre')
        predicted=model.predict(token_list,verbose=0)
        predicted_word_index=np.argmax(predicted,axis=-1)[0]
        predicted_word=tokenizer.index_word[predicted_word_index]
        seed_text += " " + predicted_word

    return seed_text
```

```python
print(predict_next_word("machine learning"))
```

```python


import matplotlib.pyplot as plt
import numpy as np
# Example input sentence to generate predictions for
input_sentence = "data Science is"

# Tokenize and pad the input sentence
token_list = tokenizer.texts_to_sequences([input_sentence])[0]
token_list = pad_sequences([token_list], maxlen=max_sequence_length - 1, padding='pre')

# Get the model's raw predictions (probabilities for each word in the vocabulary)
predictions = model.predict(token_list, verbose=0)[0]

# Get the top N predicted word indices and their confidence scores
top_n = 10 # You can adjust this number
top_indices = np.argsort(predictions)[::-1][:top_n]
top_confidences = predictions[top_indices]

# Get the actual words corresponding to the top indices
top_words = [tokenizer.index_word[index] for index in top_indices]



# Create a list of dictionaries for easier plotting with seaborn
heatmap_data = [{'word': word, 'confidence': confidence} for word, confidence in zip(top_words, top_confidences)]

# Convert to a pandas DataFrame for easier manipulation (optional but common)
import pandas as pd
heatmap_df = pd.DataFrame(heatmap_data).sort_values(by='confidence', ascending=False)

# Create a heatmap
plt.figure(figsize=(12, 3))
sns.heatmap(heatmap_df.set_index('word').T, annot=True, fmt=".2f", cmap="viridis", linewidths=.5, cbar_kws={'label': 'Prediction Confidence'})
plt.title(f'Top {top_n} Next Word Prediction Confidences for: "{input_sentence}"')
plt.yticks([]) # Hide y-axis labels if only one row
plt.xlabel('Predicted Word')
plt.show()
```
